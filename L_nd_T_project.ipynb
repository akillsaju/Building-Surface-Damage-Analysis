```python
import cv2
import numpy as np
import os

IMAGE_FOLDER = "images/"
RESULT_FOLDER = "results/"
THRESHOLD_DENSITY = 0.02

if not os.path.exists(RESULT_FOLDER):
    os.makedirs(RESULT_FOLDER)

print("Processing Images...\n")

for filename in os.listdir(IMAGE_FOLDER):

    if filename.endswith(".jpg") or filename.endswith(".png"):

        image_path = os.path.join(IMAGE_FOLDER, filename)
        image = cv2.imread(image_path)

        if image is None:
            continue

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(gray, (5, 5), 0)
        edges = cv2.Canny(blur, 50, 150)

        kernel = np.ones((3, 3), np.uint8)
        dilated = cv2.dilate(edges, kernel, iterations=1)

        crack_pixels = np.sum(dilated == 255)
        total_pixels = dilated.shape[0] * dilated.shape[1]
        defect_density = crack_pixels / total_pixels

        if defect_density < THRESHOLD_DENSITY:
            classification = "Minor Damage"
        else:
            classification = "Major Damage"

        print(f"Image: {filename}")
        print(f"Defect Density: {defect_density}")
        print(f"Classification: {classification}")
        print("--------------------------------")

        output_image = image.copy()
        cv2.putText(output_image,
                    f"{classification} | Density: {round(defect_density,4)}",
                    (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.8,
                    (0, 0, 255),
                    2)

        cv2.imwrite(os.path.join(RESULT_FOLDER, filename), output_image)

print("All Images Processed Successfully!")
```
